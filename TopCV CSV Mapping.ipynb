```python
import pandas as pd
import numpy as np
import json
import os
import itertools
```

```python
DATA_DIR = "data/topcv/json"
```

```python
def read_json(data_dir, filename):
    with open(f'{data_dir}/{filename}', 'r') as openfile:
        return json.load(openfile)
```

```python
topcv_jsons = []
for file in os.listdir(DATA_DIR):
    if file.startswith("top_cv"):
        topcv_jsons.append(read_json(DATA_DIR, file))
```

```python
industries = read_json(DATA_DIR, "industries_raw.json")
```

```python
inds_df = pd.DataFrame.from_records(industries)
inds_df['index'] = range(0, inds_df.shape[0])
inds_df
```

```python
len(topcv_jsons)
```

```python
all_jobs = {}
```

```python
for index,array16 in enumerate(topcv_jsons):
    all_jobs[index] = []
    for array in array16:
        for job in array:
            all_jobs[index].append(job)
```

```python
count_jobs = 0
max_job = 0
for key in all_jobs.keys():
    count_jobs += len(all_jobs[key])
    if len(all_jobs[key]) > max_job:
        max_job = len(all_jobs[key])
print("Total jobs: ", count_jobs)
print("Max job counts: ", max_job)
```

```python
all_jobs[2]
```

Assigning industries

```python
for i in range(0, len(all_jobs)):
    for j in range(0, len(all_jobs[i])):
        all_jobs[i][j]['industry_index'] = i
```

```python
all_jobs[2][0]
```

```python
listlist = list(all_jobs.values())
all_jobs_ls = list(itertools.chain(*listlist))
```

```python
all_jobs_df = pd.DataFrame.from_records(all_jobs_ls)
```

```python
all_jobs_df
```

```python
# all_jobs_df.to_csv("data/topcv/csv/topcv_raw.csv", index=False)
```

```python
modified_addr = []
for addr_ls in all_jobs_df['job_addresses']:
    temp_ls = []
    for addr in addr_ls:
        new_addr = addr.split(':')[0].replace('-','').strip()
        if (new_addr not in temp_ls) and not(new_addr.startswith('...')) and not('\n' in new_addr):
            temp_ls.append(new_addr)
    modified_addr.append(', '.join(temp_ls))
```

```python
modified_addr
```

```python
all_jobs_df['mod_addr'] = modified_addr
```

```python
all_jobs_df.to_csv("data/topcv/csv/topcv_raw.csv", index=False)
```
